```python
import requests as req

url_target = 'https://opentutorials.org/course/1'

res = req.get(url_target)

```

```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(res.text, "html.parser") #html 문자열과 html을 해석하는 것을 인수로 넣는다.
```

```python
soup.find_all(class_="comment_content") #class는 약어이기 때문에 class_로 써야한다.
```

```python
soup.select('#comment225751 > div.comment_content') #헤당되는 값을 모두 가져온다. 왠만하면 selector를 사용하자.
```

```python
e = soup.select('#comment > div.comment_content')[0]
e.text.strip() #문자열 앞 뒤 쓸데없는 것 지우기
```

```python
li_comment = soup.select('div.comment_content')

li_comment_processed = [ e.text.strip() for e in li_comment] #한줄로 쓰는 조건문
```

<h1> 간략한 판다스 </h1>

```python
import pandas as pd

context = {
    'comments': li_comment_processed,
    'comments2': li_comment_processed,
}
pd.DataFrame(context)
```
